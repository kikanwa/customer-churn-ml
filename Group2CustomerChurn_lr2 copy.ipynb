```python
#making use of logisitc regression as model of choice
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split


model = LogisticRegression()

#training model with data
model.fit(x[['Tenure in Months']], Y)


coefficients = model.coef_
intercept = model.intercept_

print(f"Coefficients: {coefficients}")
print(f"Intercept: {intercept}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generating values for 'Tenure in Months' for the plot
x_values = np.linspace(x['Tenure in Months'].min(), x['Tenure in Months'].max(), 100).reshape(-1, 1)

# using logisitcs regression to predict probablities 
probs = model.predict_proba(x_values)[:, 1]

# Plotting the logistic curve
plt.figure(figsize=(10, 6))
plt.scatter(x['Tenure in Months'], Y, color='black', label='Actual data')
plt.plot(x_values, probs, color='red', label='Logistic Regression Curve')
plt.title('Logistic Regression Model')
plt.xlabel('Tenure in Months')
plt.ylabel('Probability of Churn')
plt.legend()
plt.show()
```

```python
test = pd.read_csv('Group2_testing_data.csv')
```

```python
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Removing the target variable from the testing dataset
test_data_no_target = test.drop('Customer Status', axis=1)

# Using the logistic regression model to make predictions
test_predictions = model.predict(test_data_no_target[['Tenure in Months']])

# Seeing/evaluatiing the model's perfromance
accuracy = accuracy_score(test['Customer Status'], test_predictions)
conf_matrix = confusion_matrix(test['Customer Status'], test_predictions)
class_report = classification_report(test['Customer Status'], test_predictions)

print(f"Accuracy: {accuracy}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(class_report)
```

```python
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Assuming you have predictions and true labels
predictions = model.predict_proba(x[['Tenure in Months']])[:, 1]
fpr, tpr, _ = roc_curve(Y, predictions)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.plot(fpr, tpr, label=f'ROC curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Random')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()
```
